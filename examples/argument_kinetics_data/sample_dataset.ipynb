# Implementation of initial_concentrations

```python
%run ../../dfbapy/dfba.py
dfba = dFBA(
    model_path,
    verbose = False,
    printing = True,
    jupyter = True
)
dfba.simulate(
    total_time = 20,
    timestep = 5,
    initial_concentrations = initial_concentrations,
    kinetics_data = kinetics_data,
    included_metabolites = [
        'Acetate',
        'Glyceraldehyde 3-phosphate',
        'D-Fructose 6-phosphate'
    ]
)
dfba.export('simple-dfba')
```
